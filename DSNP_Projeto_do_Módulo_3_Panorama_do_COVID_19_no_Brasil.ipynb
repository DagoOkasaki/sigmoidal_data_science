#  COVID-19: OMS declara fim da Emergência de Saúde Pública de Importância Internacional

Em 4 de maio de 2023 foi realizado o 15° Encontro do Comitê de Emergência do Regulamento Sanitário Internacional sobre a pandemia do Coronavírus 2019 (COVID-19) em Genebra, Suiça. Este encontro, que vem ocorrendo periodicamente desde março de 2020, tem o objetivo de analisar a evolução da doença pelos países do globo e definir as recomendações sanitárias.

Durante esta sessão do comitê, os membros destacaram os seguintes pontos:

*   Tendência de queda nas mortes por COVID-19;
*   Queda nas hospitalizações e admissões na UTI devido à COVID-19;
*   Altos níveis de imunidade da população ao vírus SARS-CoV-2.

Com base nessas informações, o comitê aconselhou que é o momento de fazer uma transição do modo de emergência para um gerenciamento de longo prazo da COVID-19.

O Diretor Geral da Organização mundial da Saúde (OMS), Dr Tedros Adhanom Ghebreyesus, concordou com as recomendações do comitê e determinou que a COVID-19 se tornou um problema de saúde estabelecido e em andamento e que não constitui mais uma emergência de saúde pública de importância internacional (ESPII).

## O que é a COVID-19?

COVID-19 é uma doença infecciosa descoberta no final de 2019 e causada pelo vírus SARS-CoV-2.

O vírus pode ser transmitido principalmente por gotículas da boca ou nariz de pessoas infectadas quando estas tossem, espirram, cantam ou respiram.

A maioria das pessoas infectadas pelo vírus vão apresentar apenas sintomas leves a moderados de doenças respiratórias e se recuperar sem nenhum tratamento especial. Entretando, algumas vão desenvolver a forma grave da doença e necessitar de atenção médica. Pessoas de mais idade ou com comorbidades como doenças cardíacas, diabetes, doenças respiratórias crônicas ou câncer estão mais sujeitas a desenvolver a forma grave da doença. Importante ressaltar que qualquer pessoa pode ficar doente com a COVID-19 e desenvolver a forma grave da doença com risco de morte independente de sua idade.

No início da pandemia, era estimado que 15% das pessoas infectadas desenvolveriam infecções graves e precisariam de tratamento hospitalar e oxigênio. As estimativas mais recentes, [segundo a OMS](https://www.who.int/emergencies/diseases/novel-coronavirus-2019/question-and-answers-hub/q-a-detail/coronavirus-disease-covid-19#:~:text=At%20the%20start,targeted%20medical%20treatments.), sugerem que a hospitalização é necessária em torno de 3% das pessoas infectadas. Isto é graças a imunização, mutações no próprio vírus e na disponibilidade de tratamentos médicos.

Com esse recente cenário da doença e junto com a boa notícia divulgada pela OMS vamos realizar uma comparação entre os pontos destacados pelos membros do comitê e os dados públicos registrados ao longo dos mais de 3 anos de pandemia. A análise consiste em verificar se chegaremos às mesmas conclusões dos membros do comitê a nível mundial. E trazendo para uma realidade mais próxima, analisar os dados a nível Brasil para visualizar a nossa situação em relação à COVID-19.

Dependendo do volume de dados ausente, nossa análise pode ficar comprometida e apontar conclusões que não refletem a realidade. Vamos verificar a porcentagem de dados ausentes nas colunas.

No presente momento, vamos verificar qual a data mais atual disponível para seguirmos com a análise.

```
updated_date = df.date.iloc[-1]
print("Data mais recente disponível: ", updated_date.date().strftime("%d/%m/%Y"))
```

Para a data mais recente disponível, vamos seguir com a análise e verificar quais os países com mais casos e mortes.
Como o DataFrame possui também uma consolidação por regiões, vamos primeiro criar um novo DataFrame apenas com os países isolados para obtermos as informações adequadas.

```
# Filtrar as entradas referentes à regiões do DataFrame
df_regioes = df[df["iso_code"].str.contains("OWID")]
df_regioes
```

```
# Novo Dataframe a partir do original e eliminando as entradas referentes à regiões.
df_paises = df.drop(df_regioes.index)
df_paises
```

Agora que temos um DataFrame apenas dos países isolados, vamos iniciar a análise exploratória sobre estes dados.
A primeira informação que queremos obter é a contagem de casos e fatalidades para a data mais recente disponível.

```
# Países com mais casos para a data mais recente disponível
print("Data exibida: ",updated_date.date().strftime("%d/%m/%Y"),"\n")
print("Ranking dos países com mais casos: \n")
df_paises.loc[df.date == updated_date].sort_values(by="total_cases", ascending=False).head(10)
```

```
# Países com mais mortes para a data mais recente disponível
print("Data exibida: ",updated_date.date().strftime("%d/%m/%Y"),"\n")
print("Ranking dos países com mais mortes: \n")
df_paises.loc[df.date == updated_date].sort_values(by="total_deaths", ascending=False).head(10)
```

O país que lidera os rankings tanto de casos e de mortes é os Estados Unidos.

A China, que só recentemente se abriu ao mundo, viu seu número de casos explodir e se posiciona agora em segundo lugar no número de casos, mas não muito distante dos Estados Unidos na data desta análise (19/06/23).

Em terceiro lugar na quantidade de casos temos a Índia, seguida por dois países europeus: França e Alemanha.

Em relação às mortes por COVID-19, apenas 2 países se repetem em relação ao número de casos dos primeiros 5: Estados Unidos e Índia.

Após os Estados Unidos, temos Brasil em segundo, Índia em terceiro seguidos por Rússia e México.

Apenas em números é difícil de mensurar como a doença foi agressiva no início quando não tínhamos vacinas disponíveis. Vamos plotar um gráfico de barras dos 5 países com mais mortes já citados e observar a severidade da situação.


```
# Plotar um gráfico de barras para os 5 países com mais mortes
total_deaths_top5 = df_paises.loc[df.date == updated_date, ['location', 'total_deaths']].sort_values(by="total_deaths", ascending=False)[0:5]

fig, ax = plt.subplots(figsize = (10,5))
ax.ticklabel_format(style='plain')
ax.bar('location', 'total_deaths', data=total_deaths_top5)
ax.set_title('TOP 5 países com mais mortes pela COVID-19')
ax.set_xlabel('País')
ax.set_ylabel('Quantidade de mortes')
plt.grid(axis='y')


plt.tight_layout()
```

Como divulgado pela OMS, a COVID atinge um percentual da população com serveridade e os 5 países listados com mais mortes fazem parte dos países com as maiores populações mundiais.
Recentemente a Índia ultrapassou a China e tornou-se o país mais populoso do planeta, conforme divulgado pelas Nações Unidas neste [link](https://www.un.org/development/desa/dpad/publication/un-desa-policy-brief-no-153-india-overtakes-china-as-the-worlds-most-populous-country/). Em terceiro lugar estão os Estados Unidos.

A quantidade de mortes está relacionada com a quantidade de casos da doença. Vamos verificar a trajetória do número de casos e número de mortes no mundo durante todo período registrado no *DataFrame*.

```
# Plotar gráficos de linhas para ver a evolução do total de casos e mortes por COVID-19 no mundo.

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

# Segmentação do DataFrame em evolução de casos e de mortes
evolucao_casos = df.loc[df.location == "World", "total_cases"].reset_index(drop=True)
evolucao_mortes = df.loc[df.location == "World", "total_deaths"].reset_index(drop=True)

# Parametrização do primeiro gráfico
ax1.ticklabel_format(style='plain')
ax1.set_title('Evolução diária de casos de COVID-19 no mundo')
ax1.set_xlabel('Dias desde o início da Pandemia')
ax1.set_ylabel('Casos no mundo')
ax1.grid(color='grey', linestyle='-', linewidth=0.5)
ax1.plot(evolucao_casos)

# Parametrização do segundo gráfico
ax2.ticklabel_format(style='plain')
ax2.set_title('Evolução diária de mortes por COVID-19 no mundo')
ax2.set_xlabel('Dias desde o início da Pandemia')
ax2.set_ylabel('Mortes')
ax2.plot(evolucao_mortes)
ax2.grid(color='grey', linestyle='-', linewidth=0.5)

plt.tight_layout()
```

De acordo com o encontro realizado pelo comitê realizado em 4 de maio e juntamente com os gráficos acima, podemos confirmar que realmente existe uma **tendência de queda** nas mortes por COVID-19. A linha do gráfico de evolução diária de mortes está tendendo a seguir na horizontal, o que representa que o número acumulado de fatalidades diárias não está aumentando com a força que vinha anteriormente.

Ainda seguindo com a análise de mortes por COVID-19, um ponto interessante que podemos analisar com os dados que temos disponíveis é o número de mortes em relação ao PIB per capita por país. PIB per capita é um indicador que representa o grau de desenvolvimento de um país ou região e é calculado através da divisão da renda nacional pelo número de habitantes. Quanto maior este valor significa que o país é mais desenvolvido.

Vamos verificar o PIB per capita dos países com maior número de mortes e na sequencia analisar atráves de um gráfico de dispersão.


```
# Plotar um gráfico de dispersão considerando gdp_per_capita (ou seja, o PIB dos países) pelo número total de mortos. Pegue a data mais recente apenas.
gdp_deaths = df_paises.loc[df.date == updated_date, ["location", 'gdp_per_capita', 'total_deaths']].sort_values(by="total_deaths", ascending=False).dropna()
print('Dados relacionando o PIB per capita e o total de mortes em cada país:\n')
gdp_deaths.head(10)
```

```
# Plotar um Scatter plot (gráfico de dispersão) relacionando número de mortes e GDP per capita.
fig, ax = plt.subplots(figsize = (10,5))

ax.ticklabel_format(style='plain')
ax.set_title('Distribuição do total de mortes pelo PIB per capita dos países')
gdp_deaths.plot.scatter(x="gdp_per_capita", y="total_deaths", ax=ax, s=10)
ax.set_xlabel('PIB per Capita em USD')
ax.set_ylabel('Total de mortes')
ax.grid(color='grey', linestyle='-', linewidth=0.5)

plt.show()
```

### Análise Exploratória para o Brasil

Segundo os dados mais recentes, o Brasil tem destaque no cenário mundial em relação à pandemia de COVID-19 por ocupar a segunda posição no ranking de fatalidades.
Isto chama a atenção pois ocupamos apenas a sexta posição em número de casos como pudemos observar a partir dos dados analisados.

```
# Gerar um rank dos países pelo número de casos através do método rank()
ranking_casos = df_paises.loc[df.date == updated_date, ["location", 'total_cases']]
ranking_casos['rank'] = ranking_casos['total_cases'].rank(ascending=False)

# Verificar qual a posição do Brasil no rank do número de casos
ranking_casos_brasil = ranking_casos.loc[ranking_casos.location == 'Brazil', 'rank'].values[0]
print(f'O Brasil ocupa atualmente o {ranking_casos_brasil:.0f}° lugar no número de casos de COVID-19 entre todos os países')
```

Os primeiros registros de casos de COVID-19 no mundo datam do final de 2019, ainda contidos na China. Mas, devido à globalização e intensa movimentação de pessoas pelo mundo principalmente por aviões, o vírus encontrou uma forma eficiente de se espalhar por diversos países rapidamente.
O Brasil foi incluído na lista de países com casos em 2020, pouco antes da COVID-19 ser decretada como pandemia mundial em março do mesmo ano.

Vamos extrair estas informações do nosso DataFrame.

```
# Criar uma cópia do DataFrame apenas com os dados referentes ao Brasil
df_brasil = df.loc[df.location == "Brazil"].copy()
df_brasil
```

Utilizando os dados isolados referentes ao Brasil, vamos explorar as informações disponíveis.

Primeiro, vamos extrair as datas do primeiro caso registrado e também da primeira morte por COVID-19 e verificar quantos dias se passaram entre eles.

```
# Localizar o primeiro caso registrado no DataFrame
first_case_br = df_brasil.index[df_brasil["new_cases"] != 0.0][0]
print("Data do primeiro caso registrado: ", df_brasil.date.loc[first_case_br].date().strftime("%d/%m/%Y"))

# Localizar a primeira morte registrada no DataFrame
first_death_br = df_brasil.index[df_brasil["new_deaths"] != 0.0][0]
print("Data da primeira morte registrada: ", df_brasil.date.loc[first_death_br].date().strftime("%d/%m/%Y"))
```

```
# Identificar quantos dias se passaram para registrarmos a primeira morte, considerando o primeiro caso de COVID-19 registrado.
delta = first_death_br - first_case_br
print("A primeira morte por COVID-19 registrada ocorreu {} dias após o primeiro caso confirmado.".format(delta))
```

A rapidez com que a COVID-19 se espalhou pelos diversos países no começo da pandemia e na sequencia o grande número de hospitalizações e mortes deixou o mundo inteiro em estado de alerta máximo. Ninguém podia prever a situação no dia seguinte, apenas se isolar e acompanhar.
